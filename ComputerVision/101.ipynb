```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from skimage.filters import gaussian
from skimage.feature import canny 
from skimage.color import rgb2gray
from scipy.signal import correlate2d

fig = plt.figure(figsize=(18,6))

ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

ax2.set_title('Intensity')
coffee = rgb2gray(data.coffee())
print(len(coffee[0:20]))
print(len(coffee[40:60]))
coffee = coffee[140:150, 120:140]
print(coffee)
ax1.imshow(coffee, plt.cm.gray)
ax2.plot(coffee[0, 0:20])

plt.tight_layout()
plt.show()
```

# adding two images together

```python

fig = plt.figure(figsize=(18,6))

ax1 = plt.subplot(1,2, 1)
ax2 = plt.subplot(1,2, 2)

ax1.set_title('Coffee')
ax2.set_title('Coins')
coins = rgb2gray(data.coins()[0:300, 0:300])
coffee = rgb2gray(data.coffee()[0:300, 0:300])
print(coins[0:10, 0:10])
print('coffee')
print(coffee[0:10, 0:10])


concated = np.concatenate((coffee, coins), axis=1)

ax1.imshow(coffee, plt.cm.gray)
ax2.imshow(concated, plt.cm.gray)
plt.tight_layout()
plt.show()
```
